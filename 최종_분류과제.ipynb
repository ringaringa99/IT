```python
health_sample = health.loc[:,['치아우식증유무','허리둘레','LDL콜레스테롤','식전혈당(공복혈당)','체중(5Kg 단위)']]

health_sample.dropna(inplace=True)
```

#의사결정나무모델

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X = health_sample.iloc[:, 1:] 
y = health_sample['치아우식증유무']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=42)

tree = DecisionTreeClassifier(random_state=0)  
tree.fit(X_train, y_train)
print("학습용 데이터 정확도: {:.3f}".format(tree.score(X_train, y_train)))
print("시험용 데이터 정확도: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train, y_train)

print("학습용 데이터 정확도: {:.3f}".format(tree.score(X_train, y_train)))
print("시험용 데이터 정확도: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
from sklearn.tree import export_graphviz
import graphviz

export_graphviz(tree, out_file="tree.dot", class_names=[ "Yes", "No"],
                feature_names=health_sample.columns[1:], impurity=False, filled=True)

with open("tree.dot") as f:
    dot_graph = f.read()

display(graphviz.Source(dot_graph))
```

```python
result = tree.feature_importances_
print("변수의 중요도:\n", result)
```

왼쪽부터 '허리둘레','LDL콜레스테롤','식전혈당(공복혈당)','체중(5Kg 단위)' 순서

```python
import numpy as np

def plot_feature_importances_wisc(model):
    plt.rcParams["figure.figsize"] = (9,6)
    n_features = len(health_sample.columns[1:])  
    plt.barh(np.arange(n_features), model.feature_importances_, align='center') 
    plt.yticks(np.arange(n_features), health_sample.columns[1:])   
    plt.xlabel("feature importance")
    plt.ylabel("feature")
    plt.ylim(-1, n_features) 

plot_feature_importances_wisc(tree)
```

위에서부터 체중(5kg 기준), 식전혈당(공복혈당), LDL콜레스테롤, 허리둘레 순

```python
res = pd.DataFrame(result)
res["name"] = health_sample.columns[1:]
display(res.info())

display(res.sort_values(by='name'))  
```

#의사결정나무 결과
위의 의사결정나무 분류모델을 통해 치아우석증유무의 여부에 미치는 영향은 LDL콜레스테롤, 식전혈당, 허리둘레, 체중(5kg 단위) 중 공통적으로 LDL콜레스테롤, 식전혈당이 가장 큰 영향을 미치는 것으로 나타났다. 정확도는 높은 수치를 기록하긴 하지만 수치의 차이는 변수별로 크지 않아 유의미하다고 보기는 어려울 것 같다.

#랜덤포레스트

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators=5, random_state=2)
forest.fit(X_train.iloc[:, 0:2], y_train)

fig, axes = plt.subplots(2, 3, figsize=(20, 10))
for i, (ax, tree) in enumerate(zip(axes.ravel(), forest.estimators_)):
    ax.set_title("Tree {}".format(i))
    mglearn.plots.plot_tree_partition(X.iloc[:, 0:2].values, y, tree, ax=ax)
    
mglearn.plots.plot_2d_separator(forest, X.iloc[:, 0:2].values, fill=True, ax=axes[-1, -1], alpha=.4)
axes[-1, -1].set_title("Random forest")
mglearn.discrete_scatter(X.iloc[:, 0], X.iloc[:, 1], y)
plt.show()
```

```python
forest = RandomForestClassifier(n_estimators=100, random_state=0)
forest.fit(X_train, y_train)

print("학습용 데이터 세트 정확도: {:.3f}".format(forest.score(X_train, y_train)))
print("시험용 데이터 세트 정확도: {:.3f}".format(forest.score(X_test, y_test)))
```

```python
plot_feature_importances_wisc(forest)
```

랜덤포레스트의 결과 역시 의사결정나무와 동일한 것을 확인할 수 있다.

```python
from sklearn.model_selection import cross_val_score

dt_scores = cross_val_score(tree, X_train, y_train, cv=10, scoring='accuracy')  
rf_scores = cross_val_score(forest, X_train, y_train, cv=10, scoring='accuracy')

print("Accuracy")
print("Decision tree: ", dt_scores) 
print("Random forest: ", rf_scores) 

print("Accuracy mean")
print("Decision tree :{:.3f}".format(dt_scores.mean())) 
print("Random forest :{:.3f}".format(rf_scores.mean())) 
```

교차검증을 10번 한 결과 모두 평균에서 크게 벗나지 않고 비슷한 결과를 띔을 확인할 수 있다. 

```python
cv_list = [    
            ['decision_tree',dt_scores],
            ['random_forest',rf_scores],
          ]
health = pd.DataFrame.from_dict(dict(cv_list))
health.plot()
```

어떠한 변수가 미치는 영향이 가장 큰지에 대한 결과는 거의 비슷하게 나타났으나, 실제 정확도는 차이를 보인다. 랜덤포레스트가 의사결정나무에 비해 약 0.1에서 0.15정도 가량의 높은 정확도를 보인다고 위의 그래프가 설명하고 있다.

#교차검증

```python

dt_scores = cross_val_score(tree, X_train, y_train, cv=10, scoring='accuracy')  
rf_scores = cross_val_score(forest, X_train, y_train, cv=10, scoring='accuracy')
gbrt_scores = cross_val_score(gbrt, X_train, y_train, cv=10, scoring='accuracy')

print("Accuracy")
print("Decision tree: ", dt_scores) 
print("Random forest: ", rf_scores) 
print("Gradient boosting: ", gbrt_scores) 

print("Accuracy mean")
print("Decision tree :{:.3f}".format(dt_scores.mean())) 
print("Random forest :{:.3f}".format(rf_scores.mean())) 
print("Gradient boosting :{:.3f}".format(gbrt_scores.mean())) 
```

```python
from sklearn.model_selection import StratifiedKFold
kf = StratifiedKFold(n_splits=10)

print(kf)

dt_scores = cross_val_score(tree, X_train, y_train, cv=kf, scoring='accuracy')  
rf_scores = cross_val_score(forest, X_train, y_train, cv=kf, scoring='accuracy')
gbrt_scores = cross_val_score(gbrt, X_train, y_train, cv=kf, scoring='accuracy')

print("Accuracy mean -- StratifiedKFold 이용")
print("Decision tree :{:.3f}".format(dt_scores.mean())) 
print("Random forest :{:.3f}".format(rf_scores.mean())) 
print("Gradient boosting :{:.3f}".format(gbrt_scores.mean())) 
```

#인사이트 도출

의사결정나무, 랜덤포레스트, 그래이언트 부스팅 분석기법을 통해 건강검진 결과 데이터상에서 치아우식증여부, 즉 충치와 같은 질병에 어떠한 변수가 가장 큰 영향을 주는지에 대해 알아보았다. 이 때, 필요한 변수를 치아우식증 여부에 영향이 있을 것으로 예상되는 것들로 선정했다. 우선 치아 우식증은 치아의 법랑질이 손상되어 그 아래 상아질과 치수에 충치가 생기는 질환을 뜻한다. 서울아산병원 질환백과에 따르면, 플라크를 이루고 있는 세균에 의해 입안에 남아 있는 설탕, 전분 등이 분해되면, 이로 인해 발생하는 산이 치아면의 법랑질을 공격하여 손상을 입혀서 치아 우식증이 발생하며, 당뇨 환자의 경우 혈당이 잘 조절되지 못하면 치아 우식증(충치)가 잘 발생한다고 한다.(출처:https://www.amc.seoul.kr/asan/healthinfo/disease/diseaseDetail.do?contentId=32462) 따라서 식전혈당(공복혈당) 지수가 여부에 큰 영향을 주지 않을까 예상했고, 앞서 언급됐듯이 단 음식을 많이 섭취할 경우 생길 가능성이 크다고 생각해 당을 많이 섭취함으로서 야기되는 비만과 관련된 변수인 허리둘레, 체중, LDL콜레스테롤까지 변수로 설정했다. 네 변수 모두 클 수록 부정적인 결과를 낳는, 비례적인 성격을 띄기 때문에 따로 파생변수를 설정하지 않았다.  

결과적으로 의사결정나무와 랜덤포레스트 공통적으로 LDL콜레스테롤이 가장 영향력이 큰 것으로 분석되었다. 그리고 교차검증을 통해 셋 중 가장 정확한 것으로 판명된 그래디언트분석에서는 허리둘레가 가장 중요한 것으로 나타났다. 

우선 결과가 다르게 도출된 것에 대한 의문이 들 수 있으나, 변수간의 눈에 띄는 결과의 차이는 없어서 어느 한 변수가 미치는 영향이 절대적이라고 판단할 수는 없었다. 하지만 공통적으로 체중에 비해 나머지 세 변수가 큰 비율을 차지하고 있었기에, 사람의 키나 근육량 등 여러 요인에 의해 차이가 큰 체중보다는 수치가 높을수록 건강에 직결되는 영향을 주는 변수들이 더 치아우식증 여부에 영향을 준다는 것을 결과를 통해 알 수 있었다. 또, 사전 조사를 통해 공복혈당이 가장 큰 영향을 미치지 않을까 예상했었는데 예상과는 다른 결과를 얻게 되었다.
